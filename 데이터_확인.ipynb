```python
import pandas as pd
import numpy as np
from pathlib import Path

# 현재 작업 폴더 기준
BASE_DIR = Path(".")

# 파일 경로
train_path = BASE_DIR / "hanwoo_train" / "hanwoo_train.csv"
weather_path = BASE_DIR / "hanwoo_weather.csv"
area_path = BASE_DIR / "hanwoo_area.csv"
death_path = BASE_DIR / "hanwoo_death.csv"
lineage_path = BASE_DIR / "hanwoo_lineage" / "hanwoo_lineage.csv"
```

```python
# 한우 도체 등급 판정 데이터
train = pd.read_csv(train_path)

# 기상 데이터
weather = pd.read_csv(weather_path)

# 농장 사육현황 및 면적 데이터
area = pd.read_csv(area_path)

# 농장별 폐사 데이터
death = pd.read_csv(death_path)

# 개체 혈통 데이터
lineage = pd.read_csv(lineage_path)
```

## 결측치 확인 


```python
import pandas as pd

files = {
    "hanwoo_area": "hanwoo_area.csv",
    "hanwoo_weather": "hanwoo_weather.csv",
    "hanwoo_death": "hanwoo_death.csv",
    "hanwoo_train": "hanwoo_train/hanwoo_train.csv",
    "hanwoo_lineage": "hanwoo_lineage/hanwoo_lineage.csv"
}

for name, path in files.items():
    print("=" * 50)
    print(name)

    df = pd.read_csv(path)

    # 진짜 빈값 개수
    missing_count = df.isna().sum().sum()
    shape = df.shape

    # -99 개수
    minus99_count = (df.astype(str) == "-99").sum().sum()
    print("행,열 개수:", shape)
    print("빈값 개수:", missing_count)
    print("-99 개수:", minus99_count)
    print("빈값 + -99 개수:", missing_count + minus99_count)
```
